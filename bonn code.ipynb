### Bonn dataset: 
                set A patient-  'Z' file - NONSEIZURE
                set E patient-  'S' file - SEIZURE

```python
import numpy as np
import pandas as pd
nonseizure=[]
for i in range(1,10):
    df=pd.read_csv('Z'+'00'+str(i)+'.txt',header=None)
    nonseizure.append(df)
    result1 = pd.concat(nonseizure,axis=1)

for i in range(10,100):
    df=pd.read_csv('Z'+'0'+str(i)+'.txt',header=None)
    nonseizure.append(df)
    result1 = pd.concat(nonseizure,axis=1)
    

df=pd.read_csv('Z100.txt',header=None)
nonseizure.append(df)
result1 = pd.concat(nonseizure,axis=1)
    
    
```

```python
import numpy as np
import pandas as pd
seizure=[]
for i in range(1,10):
    df=pd.read_csv('S'+'00'+str(i)+'.txt',header=None)
    seizure.append(df)

    result = pd.concat(seizure,axis=1)
    
for i in range(10,100):
    df=pd.read_csv('S'+'0'+str(i)+'.txt',header=None)
    seizure.append(df)

    result = pd.concat(seizure,axis=1)

df=pd.read_csv('S100.txt',header=None)
seizure.append(df)
result= pd.concat(seizure,axis=1)
    
    
```

```python
res=[result,result1]
X = pd.concat(res,axis=0)
```

```python
train_x=X.iloc[:,0:]
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(train_x)
scaler.transform(train_x)
```

```python
train_x.shape
```

### trainy

```python
ys=np.ones(4097)
yns=np.zeros(4097)
train_y=np.concatenate((ys,yns),axis=0)
train_y.shape
```

```python
from sklearn.model_selection import train_test_split
train_X,val_X,train_Y,val_Y=train_test_split(train_x,train_y,test_size=0.25)
```

# ANN MODEL

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras import layers,models
```

```python
model=Sequential()
#model.add(layers.Flatten())
model.add(Dense(16,input_dim=100,activation='relu'))
model.add(Dense(12))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history=model.fit(train_X,train_Y,epochs=50)
```

```python
pred=model.predict(val_X)
pred.round()
```

```python
from sklearn.metrics import accuracy_score,precision_score
a=accuracy_score(val_Y,pred.round())
print(a*100)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(val_Y,pred.round())
print(cm)
```

```python
confusion_df = pd.DataFrame(cm)
labels_dict = {
    0: 'non-seizure',
    1: 'seizure'}
confusion_df = confusion_df.rename(columns=labels_dict)
confusion_df.index = confusion_df.columns
confusion_df
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score

accuracy = accuracy_score(val_Y,pred.round())
print('Accuracy: %f' % accuracy)

precision = precision_score(val_Y,pred.round())
print('Precision: %f' % precision)

recall = recall_score(val_Y,pred.round())
print('Recall: %f' % recall)

f1 = f1_score(val_Y,pred.round())
print('F1 score: %f' % f1)

# ROC AUC
auc = roc_auc_score(val_Y,pred.round())
print('ROC AUC: %f' % auc)
```

```python
import matplotlib.pyplot as plt
acc = history.history['accuracy']
#val_acc = history.history['val_accuracy']
loss = history.history['loss']
#val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, label='Training accuracy')
#plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')

plt.figure()

plt.plot(epochs, loss, label='Training Loss')
#plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```
